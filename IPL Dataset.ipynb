## Find all the unique types of dismissals in the data frame.

```python
dismissal_types=ipl_df.dismissal_kind.unique()
dismissal_types
```

## Show the distribution of all the dismissals using a Pie Chart.

```python
dismissal_counts=ipl_df.dismissal_kind.value_counts()
dismissal_counts
```

```python
ax=plt.subplots(figsize=(15,10))
ax=plt.pie(dismissal_counts,labels=dismissal_counts.keys())

plt.title("Dismissal Types Distribution",fontsize=30)
plt.show()
```

## Find total runs scored by each batsman in ipl from 2008-2020

```python
ipl_df.groupby('batsman').batsman_runs.sum()
```

## Find the total no of wickets taken by each bowler in ipl from 2008-2020

```python
ipl_df.groupby('bowler').is_wicket.sum()
```

## plot the Total runs scored on each ball of the over by using bar graph.

```python
total_runs_df=ipl_df.groupby('ball').total_runs.sum()
total_runs_df
```

```python
##total_runs_df.plot(kind='bar')
plt.bar(total_runs_df.keys(),total_runs_df.values,color=['black', 'red', 'green', 'blue', 'cyan','yellow','purple','pink'])
plt.title("Total Runs on Each ball")
plt.xlabel("Balls")
plt.ylabel("Total runs")
```

## Find the extras types and runs scored by each type.

```python
extra_df=ipl_df.groupby('extras_type').extra_runs.sum()
extra_df
```

## Show the distribution of all the extra types using a Bar Graph. 

```python
plt.bar(extra_df.keys(),extra_df.values,color=['black', 'red', 'green', 'blue', 'cyan'])
plt.figure(figsize=(50,90))
```
